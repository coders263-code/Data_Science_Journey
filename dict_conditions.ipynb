# ***Dictionary***


### ***store Data in key value Format***
### ***Key should Always be Unique***
### ***Indexing for Accessing values are not allowed***

```python
dic={"name":"saadi","Class":"3B","Department":"DSAI"}
dic["Class"]
```

### ***Dictinary inside a dictionary***

```python
dic2={"name":"saadi","Class":"3B","Department":"DSAI","data":{1:"fahad",2:"umer",3:"Khadija",4:"Asma"}}
dic2["data"][2]
```

### ***Deletion of any value by key***

```python
del dic2["name"]
print(dic2)
```

### ***Adding values in dictionary***

```python
dic2["clg"]="KFUEIT"
print(dic2)
```

### ***use of .keys() & .value()***

```python
print(dic2.keys())
# print values
print(dic2.values())

#convert Any dictionary into a list
list(dic2)
```
