### ***Use of "W"***
##### ***First tuncate data,then insert new data***
##### ***Example***

```python
try:
    f=open("text.txt","w")
# f is a variable
# open is used to perform all the functions in file handling
    f.write("My name is Muhammad saad Ali")
except FileNotFoundError as e:
    print(e)
finally:
    f.close()
    print("Program has been completed")
```

### ***Use of a***
##### ***By using "a" we can insert data at the end of file without deleting Existing data***
##### ***Example***

```python
try:
    f=open("text.txt","a")
    f.write("My Father name is Miskeen Ali")
except FileNotFoundError as e:
    print(e)
finally:
    f.close()
    print("Your code has been Executed !")
```

### ***Use of r***
##### ***"r" is used to open files in read mode***
##### ***Example***

```python
try:
    f=open("text.txt","r")
    data=f.read()
    print(data)
except FileNotFoundError as e:
    print(e)
finally:
    f.close()
    print("Your code has been Executed !")
```
