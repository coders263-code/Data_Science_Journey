### ***Differ B/W print() & return in any function***
          ***print() provide only string***
          ***return give any type of data when we call any function***

```python
# example by print()
def test1():
    print("My Name is M.saad Ali")

# call function
test1()
```

```python
# Example by return
# We can perform all operation of strings by using return
def test2():
    return "My name is M.saad Ali"

test2()
```
