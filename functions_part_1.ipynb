### ***Differ B/W print() & return in any function***
          ***print() provide only string***
          ***return give any type of data when we call any function***

```python
# example by print()
def test1():
    print("My Name is M.saad Ali")

# call function
test1()
```

```python
# Example by return
# We can perform all operation of strings by using return
def test2():
    return "My name is M.saad Ali"

test2()
```

### *args
         ***by using it we can pass multiple values in a function***
        ***Example***

```python
def test4(*args):
    return args

test4(1,2,3,4,5,6)
```

### ***For obtaining key value payer format values in a functions we use double asteric test(**kwargs)***

```python
def test5(**kwargs):
    return kwargs

test5(a=[1,2,3,4,5],b="saadi")
```
