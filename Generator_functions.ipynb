##### ***Generator Function Reduce Memory Consumptions***
##### ***"yield" is keyword used to  make generater Functions***
##### ***return is Not used in Generator Functions***
##### ***Extract Data One By One***

```python
# Fibonacci Numbers
def test_fibo(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b=b,a+b

for i in test_fibo(10):
    print(i)
```
