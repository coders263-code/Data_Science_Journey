#### ***Hides Data & any implementation by double "__"*** 

```python
# Example
class Bank_Account:
    def __init__(self,Balance):
        self.__Balance=Balance
    def deposit(self,amount):
        self.__Balance=self.__Balance+amount
        return True
    def withdraw(self,amount):
        if amount<=self.__Balance:
            self.__Balance=self.__Balance-amount
            return True
        else:
            return False
        
    # This Method For Balnce checking
    def return_Balance(self):
        return self.__Balance
        
```

```python
saadi=Bank_Account(1000000)
saadi.deposit(500)
saadi.return_Balance()
```
