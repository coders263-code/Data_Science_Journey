##### ***Multiple Behaviour Of A single Entity***

```python
class Data_science:
    def syllabus(self):
        print("This is My Data science Syllabus!")
class web_dev:
    def syllabus(self):
        print("This is  my dev syllabus")

def parcer(class_obj):
    for i in class_obj:
        i.syllabus()


Data_science=Data_science()
web_dev=web_dev()

class_obj=[Data_science,web_dev]
parcer(class_obj)
```
