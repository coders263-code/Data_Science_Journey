### ***Constructor***
##### ***Self is not Reserved Keyword,we can use any word as a first argument in method,its a pointer***


```python
class Student:
    def __init__(self,Name,Roll_Number,Gmail):
        self.name=Name
        self.roll_Number=Roll_Number
        self.gmail=Gmail
    def return_data(self):
        return self.name,self.roll_Number,self.gmail

saadi=Student("saadi",89,"saadi@gmail.com")
saadi.return_data()
```
