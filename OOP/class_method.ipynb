```python
class Student:
    def __init__(self, name, age, marks):
        self._name = name        # protected variable
        self._age = age
        self._marks = marks

    # ---------- Getter ----------
    @property
    def marks(self):
        print("Getting marks...")
        return self._marks

    # ---------- Setter ----------
    @marks.setter
    def marks(self, value):
        print("Setting marks...")
        if value < 0 or value > 100:
            print("❌ Invalid marks! Marks must be between 0 and 100.")
        else:
            self._marks = value

    # ---------- Deleter ----------
    @marks.deleter
    def marks(self):
        print("Deleting marks...")
        del self._marks

    # Normal method
    def display(self):
        print("Student Name:", self._name)
        print("Student Age:", self._age)
        if hasattr(self, '_marks'):
            print("Student Marks:", self._marks)
        else:
            print("Student Marks: Deleted")


# --------- Object Creation ---------
s1 = Student("Ali", 20, 85)

# Display data
s1.display()
print()

# Using Getter
print("Marks:", s1.marks)
print()

# Using Setter
s1.marks = 95
print("Updated Marks:", s1.marks)
print()

# Invalid Setter Value
s1.marks = 150
print()

# Using Deleter
del s1.marks
s1.display()
```
