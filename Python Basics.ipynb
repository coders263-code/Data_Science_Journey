# ***Variables***


```python
## We can store Any type of data in a variable
## In python,No need to declare the type of variable
## we can fid type of variable using type() function
## Python is dynamically typed language
## Example:
a=5
b=3.14
c="Hello, World!"
d=[1, 2, 3, 4, 5]
e=(10, 20, 30)
f={'name': 'Alice', 'age': 25}
g={1, 2, 3, 4, 5}
## Now ,let's check the type of each variable
print(type(a))  # Output: <class 'int'>
print(type(b))  # Output: <class 'float'>
print(type(c))  # Output: <class 'str'>
print(type(d))  # Output: <class 'list'>
print(type(e))  # Output: <class 'tuple'>
print(type(f))  # Output: <class 'dict'>
print(type(g))  # Output: <class 'set'>
# Example of dynamic typing
x = 10          # x is an integer
print(type(x))  # Output: <class 'int'>
x = "Python"    # Now x is a string
print(type(x))  # Output: <class 'str'>
```

# ***Indexing***
           access any value in a string, we can access individual characters using indexing
           ## Example

```python
## We have String
a="saadi"
## Now we can access individual characters using indexing
print(a[0])
## Backward indexing
print(a[-1])
## Slicing
print(a[0:3])
## jumping in slicing
print(a[0:5:2])
## print(a[0:5:0]) it gives error,because step cannot b e zero in indexing
## 
```
