#### ***Tells starting and ending of any function***
#### ***To declare it by using "@"***

```python
# Example
def test_dec(fun):
    def test_inner():
        print("Starter of this function")
        fun()
        print("Ending of this function")
    return test_inner

@test_dec
def test1():
    print(6+7)

test1()
```

#### ***Timer testing by using Decorators***

```python
import time
def deco_test(func):
    def timer(limit):
        start=time.time()
        result=func(limit)
        end=time.time()
        print("Time taken:",end-start)
        return result
    return timer

@deco_test
def num(limit):
    for i in range(limit):
        print(i)
    return "Done"

print(num(10))
```
