#### ***Tells starting and ending of any function***
#### ***To declare it by using "@"***

```python
# Example
def test_dec(fun):
    def test_inner():
        print("Starter of this function")
        fun()
        print("Ending of this function")
    return test_inner

@test_dec
def test1():
    print(6+7)

test1()
```
