## ***Exception Handling***


#### ***Used to handle errors***
#### ***Save our code from any crash***
#### ***Have 4 types Of Blocks***
#### ***try,except,else,finally***
#### ***"try" block always run***
#### ***"try" block Executed succefully then "else" block Executed***
#### ***if "try" block fail to run then "except" block run***
#### ***"finally" block Executed Always***

## ***Examples***

```python
# here Except block run because try block is going  to False
try:
    f=open("saadi.text","r")
except Exception as e:
    print("Your code has any problem",e)
```

```python
# Here try bloch will b Execute
try:
    f=open("text.txt","w")
    f.write("This is my file")
    f.close()
except Exception as e:
    print("Fail to Execute",e)

```

```python
#try except & else
try:
    f=open("text.txt","r")
except  Exception as e:
    print("Fail to Execute")
else:
    f.close()
    print("Executed")
```

```python
#try,except,else &finally
# finally always Executed
#try except & else
try:
    f=open("text.txt","r")
except  Exception as e:
    print("Fail to Execute")
else:
    f.close()
    print("finally executed")
finally:
    print("Program becomes to End")
```
