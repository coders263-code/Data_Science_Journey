# ***While loop***

### ***Example***

```python
#Syntex
#1 intialization
#2 condition checking
#3 print

a=0
while a<=10:
    print(a)
    a=a+1
```

```python
# Sum up the number till some number

n=int(input("Enter your number for limit"))
start=0
counter=1
while counter<=n:
    start=start+counter
    counter=counter+1
print(start)
```
