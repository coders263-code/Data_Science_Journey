### ***Example***

```python
#Syntex
#1 intialization
#2 condition checking
#3 print

a=0
while a<=10:
    print(a)
    a=a+1
```

```python
# Sum up the number till some number

n=int(input("Enter your number for limit"))
start=0
counter=1
while counter<=n:
    start=start+counter
    counter=counter+1
print(start)
```

```python
# Factorial of any number 
n=int(input("Enter Your number"))
f=1
num=1
while num<=n:
    f=f*num
    num=num+1

print(f)
```

```python
# Fibonacci series
n=int(input("Enter Your Number"))
a,b=0,1
counter=0
while counter<n:
    print(a)
    c=a+b
    a=b
    b=c
    counter=counter+1
```

```python
# print a table 
n=int(input("Enter Your Number for table:"))
i=1
while i<=10:
    print(n,"*",i,"=",n*i)
    i=i+1
```
