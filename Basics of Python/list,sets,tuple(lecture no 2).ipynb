```python
print(l[0:3])
# Reverse Slicing
print(l[::-1])
# Print only those items that's present in even indexing
print(l[::2])
# print string that's present in a list, first two elements 
print(l[0][0:2])
# only we can merge list with list not with any other type of data type
# Example
l1=[2,3,4]
print(l+l1)
l.
```

# ***Data insertion into A list***
### ***append , extend & insert***
          append add data at the end of list  without unrapping
          extend add data at the end of  list with rapping 
          insert used to insert data at any index 

```python
# Example of append
l.append("saadi")
print(l)
# Example of extend
l.extend("saadi")
print(l)
# Example of insert
l=[1,2,3,5,6,2]
l.insert(2,"saadi")
l.insert(-1,"ali")
print(l)
```

# ***Deletion of data into a list***
### ***pop & remove***
                       pop remove to used with indexing number
                       remove used to dell any value with value 

```python
# example of pop
l2=[2,3,4,5,[2,5,6,7]]
l2.pop(3)
print(l2)
# example of remove
l2.remove(4)
print(l2)
```

```python
l3=[2,3,5,1,[1,2,3,4]]
l3[4].remove(1)
print(l3)
```

### ***A list is sortable if it has only same data types Elements*** 
### ***Here,we discuss some other functions in list like (reverse(),sort(in Ascending order or decending with reverse ==true))***

```python
l4=[1,2,3,4,6,1,9,10,4]
#sorting
l4.sort()
print(l4)
# sorting in reverse order
l4.sort(reverse=True)
print(l4)

# Reversing of list permanently
l4.reverse()
print(l4)
```
