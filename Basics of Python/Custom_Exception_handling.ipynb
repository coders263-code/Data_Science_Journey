```python
class validage(Exception):
    def __init__(self, msg):
        self.msg=msg
def validage(age):
    if age<0:
        raise validage("Entered Age is Unvalid")
    elif age>200:
        raise validage("Entered age is out of range")
    else:
        print("Age is valid")

try:
    age=int(input("Enter Your Age :"))
    validage(age)
except validage as e:
    print(e)
```
