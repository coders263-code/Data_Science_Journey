#### ***Use Always Specific Exception***


```python
# Bad Example 
try:
    z=10/0
except Exception as e:
    print(e)
```

```python
# Best Example 
try:
    z=10/0
except ZeroDivisionError as e:
    print("I'm trying to solve this Zero Division Error",e)
```

```python
# Always try to log Your Error
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    10/0
except ZeroDivisionError as e:
    logging.error("I'm trying to solve ZerDivision Error{}".format(e))
```

##### ***Always avoid to write multiple Exception handling***
##### ***Document All The Errors***
##### ***Cleanup All The Resources***
