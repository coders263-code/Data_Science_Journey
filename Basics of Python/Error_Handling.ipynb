```python
# import Error
try:
    import saadi
except ImportError as e:
    print(e)
```

```python
try:
    d={"key":"saadi","values":[1,2,3,4,5]}
    print(d["key2"])
except KeyError as e:
    print("this is ",e," error")
```

```python
# typeerror
try:
    123 + "saadi"
except TypeError as e:
    print(e)
else:
    print("Try block is Executed ")
```

```python
# Index Error
try:
    l=[1,2,3,4,5,6,7]
    print(l[9])
except IndexError as e:
    print(e)
```

```python
# FileNot Found Error
try:
    f=open("test.txt","r")
    f.read()
except FileNotFoundError as e:
    print(e)
```
