#### ***In Buffered,Big files go to into Junk***

```python
# Examples
import io
```

```python
with open("Text.txt","wb") as f:
    file= io.BufferedWriter(f)
    file.write(b"My name is Muhammad saadi")
    file.flush()
# flush() is used just like .close to closing file
```

```python
# Read file
with open("Text.txt","rb") as f:
    file=io.BufferedReader(f)
    data=file.read()
    print(data)
```
