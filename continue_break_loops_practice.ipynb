### ***Continue & break***
#### ***Continue in loops***

```python
l=["Saadi","Ali","","umer"]
for i in l:
    if i=="Ali":
        continue
    print(i)
```

#### ***break in loops***

```python
l2=["saadi","asma","umer"]
for i in l2:
    if i=="asma":
        break
    print(i)
```

### ***range()***
#### ***Used to generate data***
##### ***Syntex*** ##### ***range(start,end,step)***

```python
for i in range(0,10,2):
    print(i)
```

### ***print elements at even index number in a list***

```python
l3=[22,33,1,55,6,2,3,34,5,6,7,7,6,6,5,4,4,3,2,234,4,5,6,3,3]
for i in range(0,len(l3),2):
    print(l3[i])
```

### ***Sum of data in a list***

```python
l5=[4,3,2,1,5,66,3,2,1,8,4,3]
result=0
for i in l5:
    result=result+i
print(result)
```
